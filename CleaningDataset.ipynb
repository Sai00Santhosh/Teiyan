```python
#Importing Libraries

import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_csv('animes.csv', encoding="unicode_escape", error_bad_lines=False)
```

```python
df = df.drop(columns=["link", "img_url", "ranked", "synopsis", "popularity", "uid"])
```

```python
# Printing out year when first aired (all data accepted without errors)

def func(string):
    
    string = string[7:12] 
    
    return string

df["aired"] = df["aired"].apply(func)
```

```python
df = df[df.members > 25000]
df
```

```python
#updating csv file
df.to_csv("animes.csv", index=False)
```
